```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#convertim fisierul csv in dataframe
df = pd.read_csv("train.csv")
```

```python
#vizualizam primele 5 linii din tabel
df.head(10)
```

```python
#Facem o coloana noua, unde sa extragem daca persoana este : Mr., Mrs., Miss., Master etc.

def extract_title(text):
    res = text.split()
    final_word = [word for word in res if word in ("Miss.", "Master.", "Mr.", "Mrs.", "Ms.", "Sir.", "Dr.", "Don.", "Col.", "Rev.", "Mlle.", " Mme.", "Major.", "Mme.", "Lady.", "Capt.", "Countess.", "Jonkheer.")]
    return final_word[0] if final_word else None

df["Title"] = df["Name"].apply(extract_title)

df.head(20)

#Rev -> un fel de preoti, cel mai probabil au sansa de supravietuire 0 
#Mlle. -> Mademoiselle., rang mare, sanse mari de supravietuire
#Master -> pentru copii (baieti), probabil au sanse de supravietuire mai mari

```

```python
df.info()
```

```python
# Observam ca Embarked are 2 linii cu valori null, asa ca inspectam acele linii. solutia aleasa este sa atribuim random Embarked pentru cele 2 linii
mask = df[df["Embarked"].isnull()].index

df.loc[mask, "Embarked"] = np.random.choice(['C', 'Q', 'S'])
```

```python
df.drop("Cabin", axis = 'columns', inplace = True) 
df.head()
```

```python
df.info()
#Mai ramane sa curatam coloana Age pentru a avea setul relativ curat
```

```python
#Vreau sa vad varsta medie in functie de titlu
df_dummy = round(df["Age"].groupby(df["Title"]).mean(), 1).reset_index().set_index("Title")
df_dummy
```

```python
# O sa modific valoriile Nan cu media in functie de status social
mask_age = df["Age"].isnull()
df[mask_age]

df.loc[mask_age, "Age"] = df.loc[mask_age,"Title"].map(df_dummy["Age"]) 

df.info()
```

```python
survival_by_status = df["Title"].groupby(df["Survived"]).value_counts().reset_index().set_index("Title")
survival_by_status
```

OBSERVATII: Observam ca barbatii care nu au neaparat un statut major, au fost predispusi la a murii MAJORITAR, comparativ cu femeile si copiii. Ar fi interesant de vizualizat aceasta informatie.

```python
# Define the color palette
custom_palette = sns.color_palette("pastel")

# Plotting the bar plot
plt.figure(figsize=(10, 6))  # Increase the size of the plot
sns.barplot(data=survival_by_status, x=survival_by_status.index, y="count", hue="Survived", palette=custom_palette)

# Adding labels and title
plt.xlabel("Survived")
plt.ylabel("Count")
plt.title("Survival Count by Title")
plt.grid()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

Vreau sa vad exact acelasi lucru, dar de data asta si in functie de clasa in care se aflau

```python
survival_by_status_and_class = df.groupby(["Survived", "Pclass", "Title"]).size().reset_index(name="count").set_index("Title")
survival_by_status_and_class
```

```python
#combinam title si pclass intr-o sinura coloana
survival_by_status_and_class["Title_Pclass"] = survival_by_status_and_class.index + ", Pclass: " + survival_by_status_and_class["Pclass"].astype(str)


# Define the color palette
custom_palette = sns.color_palette("pastel")

# Plotting the bar plot
plt.figure(figsize=(12, 8))  # Increase the size of the plot
sns.barplot(data=survival_by_status_and_class, x="Title_Pclass", y="count", hue="Survived", palette=custom_palette, dodge=True)

# Adding labels and title
plt.xlabel("Title, Pclass")
plt.ylabel("Count")
plt.title("Survival Count by Title")
plt.grid()

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

# Convertim Sex in 0 -> male, 1 -> female

```python
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({"C": 0, "Q": 1, "S" : 2})
```

# 2. Execute model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
columns = ["Pclass", "Sex", "Age", "Embarked"]
X = df[columns]
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
columns = ["Pclass", "Sex", "Age", "Embarked"]
X = df[columns]
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)

from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
